```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import BertTokenizer, BertModel



```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertModel.from_pretrained("bert-base-uncased", output_attentions=True)

```

```python
def get_attention(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    attention = outputs.attentions[-1]
    return attention, inputs['input_ids']

```

```python
def plot_attention(attention, input_ids):
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
    attention = attention[0].mean(dim=0).cpu().numpy()

    attention = attention[:len(tokens), :len(tokens)]

    plt.figure(figsize=(10, 5))
    sns.heatmap(attention, xticklabels=tokens, yticklabels=tokens, cmap='coolwarm', annot=True)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.title("Attention Heatmap")
    plt.show()

```

```python
def analyze_medical_text(text):
    print("Analyzing Text:", text)
    attention, ids = get_attention(text)
    plot_attention(attention, ids)
    print("\n" + "-"*50 + "\n")
```

```python

text1 = "Patient reports persistent cough, high fever, and difficulty breathing for the past three days."
analyze_medical_text(text1)

```

```python
text2 = "Mild headache and occasional dizziness, but no fever or cough."
analyze_medical_text(text2)

```

```python
text3 = "Patient experiences sudden paralysis and slurred speech, no fever or cough."
analyze_medical_text(text3)
```
